### 1.2. Asegurar la Existencia de IDs o Claves Únicas

El siguiente paso implica chequear que no existen datos duplicados y que las claves, si existen, son únicas.

En este sentido, se podría esperar que el `Permit Number` sea único. Sin embargo, del resumen previo se puede deducir que esto no es así, ya que la cuenta es bastante superior a la cantidad de valores únicos. Es decir, existen más de 17000 permisos con el mismo `Permit Number`.

```python
data.shape
```

```python
data_ind = data.set_index("Permit Number")
```

```python
data_ind[data_ind.index.duplicated(keep=False)].sort_index()
```

Vemos que, al menos en los registros visualizados, la diferencia parece estar principalmente en la columna `Record ID`. Por otra parte, en las explicaciones del dataset propuesto por el autor describe este campo como "algún ID, no útil para esto". Dado que no agrega información, podemos suprimir esta columna de nuestro set de datos:

```python
data_ind.columns

data_ind = data_ind[['Permit Type', 'Permit Type Definition', 'Permit Creation Date',
       'Block', 'Lot', 'Street Number', 'Street Number Suffix', 'Street Name',
       'Street Suffix', 'Unit', 'Unit Suffix', 'Description', 'Current Status',
       'Current Status Date', 'Filed Date', 'Issued Date', 'Completed Date',
       'First Construction Document Date', 'Structural Notification',
       'Number of Existing Stories', 'Number of Proposed Stories',
       'Voluntary Soft-Story Retrofit', 'Fire Only Permit',
       'Permit Expiration Date', 'Estimated Cost', 'Revised Cost',
       'Existing Use', 'Existing Units', 'Proposed Use', 'Proposed Units',
       'Plansets', 'TIDF Compliance', 'Existing Construction Type',
       'Existing Construction Type Description', 'Proposed Construction Type',
       'Proposed Construction Type Description', 'Site Permit',
       'Supervisor District', 'Neighborhoods - Analysis Boundaries', 'Zipcode',
       'Location']]
data_ind.shape
```

Si borramos los duplicados, considerando los valores de todas las columnas restantes, tendremos lo siguiente:

```python
data_ind_uni = data_ind.drop_duplicates(subset = ['Permit Type', 'Permit Type Definition',
        'Permit Creation Date','Block', 'Lot', 'Street Number', 'Street Number Suffix', 'Street Name',
       'Street Suffix', 'Unit', 'Unit Suffix', 'Description', 'Current Status',
       'Current Status Date', 'Filed Date', 'Issued Date', 'Completed Date',
       'First Construction Document Date', 'Structural Notification',
       'Number of Existing Stories', 'Number of Proposed Stories',
       'Voluntary Soft-Story Retrofit', 'Fire Only Permit',
       'Permit Expiration Date', 'Estimated Cost', 'Revised Cost',
       'Existing Use', 'Existing Units', 'Proposed Use', 'Proposed Units',
       'Plansets', 'TIDF Compliance', 'Existing Construction Type',
       'Existing Construction Type Description', 'Proposed Construction Type',
       'Proposed Construction Type Description', 'Site Permit',
       'Supervisor District', 'Neighborhoods - Analysis Boundaries', 'Zipcode',
       'Location'], inplace=False)

data_ind_uni.shape
```

Vemos que sólo se han eliminado 273 registros, por lo que aún permanecen `Permit Numbers` repetidos. Analicemos nuevamente los valores de los restantes campos:

```python
data_ind_dupl = data_ind_uni[data_ind_uni.index.duplicated(keep=False)].sort_index()
# data_ind_dupl.to_csv("Output/analisis_permit_num_repetido.csv")
data_ind_dupl
```

En muchos casos, se observa que tanto `Street Number` como `Lot` pueden tener diferentes valores para el mismo `Permit Number`. Esto es algo factible, ya que el mismo permiso pudo haber aplicado a diferentes construcciones. Construimos una clave que los incorpore:

```python
data_ind_uni.reset_index(inplace = True)
```

```python
data_ind_uni['ID'] = data_ind_uni['Permit Number'] + ' - ' + data_ind_uni['Lot'] + ' - ' + data_ind_uni['Street Number'].astype(str)
data_ind_uni.head(2)
```

Una vez que obtenemos este ID, volvemos a preguntar por los valores duplicados:

```python
data_ind_uni.set_index('ID', inplace = True)
```

```python
data_ind_dupl = data_ind_uni[data_ind_uni.index.duplicated(keep=False)].sort_index()

data_ind_dupl.head(20)
```

De 31489 `Permit Numbers` duplicados, pasamos a tan solo 893 registros repetidos. Además, vemos que la principal diferencia de valores entre `ID's` repetidos se produce en las columnas `Street Number Suffix`, `Unit` y `Unit Suffix`. Para asegurarnos que tome aquellos registros que no tienen NaN en estos campos, ordenamos inicialmente el dataset.

```python
data_ind_uni.sort_values(by = ['Current Status Date','Street Number Suffix', 'Unit Suffix','Unit'], axis=0, ascending=False, inplace=True,
                         na_position='last')

data_ind_uni.tail(5)
```

```python
data_ind_uni2 = data_ind_uni.drop_duplicates(subset = ['Permit Number','Permit Type', 'Permit Type Definition',
        'Permit Creation Date','Block', 'Lot', 'Street Number', 'Street Name',
       'Street Suffix', 'Description', 'Current Status',
       'Current Status Date', 'Filed Date', 'Issued Date', 'Completed Date',
       'First Construction Document Date', 'Structural Notification',
       'Number of Existing Stories', 'Number of Proposed Stories',
       'Voluntary Soft-Story Retrofit', 'Fire Only Permit',
       'Permit Expiration Date', 'Estimated Cost', 'Revised Cost',
       'Existing Use', 'Existing Units', 'Proposed Use', 'Proposed Units',
       'Plansets', 'TIDF Compliance', 'Existing Construction Type',
       'Existing Construction Type Description', 'Proposed Construction Type',
       'Proposed Construction Type Description', 'Site Permit',
       'Supervisor District', 'Neighborhoods - Analysis Boundaries', 'Zipcode',
       'Location'])
#Not 'Street Number Suffix', 'Unit Suffix','Unit'
data_ind_uni2.shape
```

Se eliminaron 478 registros más. Repetimos el procedimiento para volver a analizar los duplicados remanentes.

```python
data_ind_dupl = data_ind_uni2[data_ind_uni2.index.duplicated(keep=False)].sort_index()
data_ind_dupl.shape
```

```python
data_ind_dupl
```

Vemos que permanecen 32 registros repetidos aún. En muchos de los casos, la diferencia en los valores se produce en las columnas `Street Name` y `Street Suffix` así también como en los campos de fechas. En el caso del nombre de la calle y sufijo, probablemente se trate de construcciones en esquinas de manzana, que dan a dos calles.
