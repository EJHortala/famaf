3. Implementar dos modelos (como ser Random Walk (Naïve) y Exponential Smoothing)  

Se opta por predecir la variable: Total de minutos salientes (miles)

## Modelo Random walk

```python
data.index[-1]+1
```

```python
y = data["Total de minutos salientes (miles)"]
```

```python
data.iloc[-1]["Total de minutos salientes (miles)"]
```

```python
h = 4
test=pd.period_range(start=data.index[-1]+1, periods=h, freq='Q')

pred=pd.Series(np.repeat(data.iloc[-1]["Total de minutos salientes (miles)"], h), index=test) # the forecast repeats the last observed values h times

pred
```

```python
resid=y-y.shift(1) # the shift lags the series by one period
sigma = resid.std()
```

```python
intv = pd.concat([pred-stats.norm.ppf(0.975)*sigma*np.sqrt(np.arange(1,h+1)),
                   pred+stats.norm.ppf(0.975)*sigma*np.sqrt(np.arange(1,h+1))], axis=1)
intv.round(3)
```

```python
point_pred = (intv[0]+intv[1])/2

fig, ax = fanchart(y['2010':],point_pred,intv,intv,intv)

plt.show()
```

Usando el modelo random walk la curva predicha es similar a la curva original desplazada una unidad, hasta el último dato conocido y a partir de allí se mantiene constante en ese valor. Esta predicción resulta confiable solo para valores de h muy pequeños ya que se observa una ampliación exponencial del intervalo de confianza para valores mayores.

## Modelo exponential smoothing

```python
model_ses = sm.tsa.SimpleExpSmoothing(data["Total de minutos salientes (miles)"]).fit()
```

```python
fitted_ses = model_ses.fittedvalues
# model_ses.summary() # summary isn't implemented in 0.9, looks it'll be avaible in 0.10
alpha_ses = model_ses.model.params['smoothing_level']
mse_ses = sm.tools.eval_measures.mse(data["Total de minutos salientes (miles)"],fitted_ses)
# loglik_ses = model_ses.model.loglike(model_ses.model.params) # NotImplementedError
aic_ses = model_ses.aic
bic_ses = model_ses.bic
print(' Smoothing parameter: alpha  {0:.3f}'.format(alpha_ses))
print(' In-sample fit:')
print(' MSE               {0:.3f}'.format(mse_ses))
print(' AIC               {0:.3f}'.format(aic_ses))
print(' BIC               {0:.3f}'.format(bic_ses))
```

```python
N = len(y)
loglik_ses=-(N/2)*(1+np.log(2*np.pi))-(N/2)*np.log(mse_ses)
aic_ses=-2*loglik_ses+2*3
bic_ses=-2*loglik_ses+np.log(N)*3
print(' Log-likelihood    %.3f' % (loglik_ses))
print(' AIC               {0:.3f}'.format(aic_ses))
print(' BIC               {0:.3f}'.format(bic_ses))
```

```python
fig, ax= plt.subplots(figsize=(15,10))
y.plot(color=red, label='minutos')
fitted_ses.plot(color='black', label='Simple exponential smoothing fit', alpha=0.8)
ax.set_xlabel('')
ax.set_ylabel('real data')
ax.set_title('Total de minutos salientes (miles)')
ax.set_xticks([], minor=True) 
plt.legend(loc='best')
sns.despine()
plt.show()
```

El modelo simple exponential smoothing muestra una curva de predicción suavizada y desplazada una unidad con respecto a los valores originales.

4. Visualizar los resultados y evaluarlos, comparando ambas implementaciones  

h = 5
w = h*2
model = sm.tsa.SimpleExpSmoothing
pred_index = y.index[-1] + np.array(range(1,h+1))
pred_rolling = pd.Series(0,index=pred_index)
pred_expanding = pd.Series(0,index=pred_index)
orig_fit = None
for i in range(0,h):
    current = pd.concat([y, pred_expanding.head(i)])
    expanding = model(current).fit()
    if i == 0:
        orig_fit = expanding.fittedvalues
    rolling = model(current.tail(w)).fit()
    pred_expanding.update(expanding.forecast(1))
    pred_rolling.update(rolling.forecast(1))
ax = y[-2*w:].plot(color='red',figsize=(10,6))
pd.concat([orig_fit[-2*w:],pred_expanding]).plot(ax=ax,color='blue')
pd.concat([orig_fit[-2*w:],pred_rolling]).plot(ax=ax,color='green')

```python
model_RW = sm.tsa.SimpleExpSmoothing(y).fit(smoothing_level=1)
fitted_RW = model_RW.fittedvalues
predic_RW = model_RW.predict(test[0],test[3])

model_RW = pd.concat([fitted_RW,predic_RW])

model_ses = sm.tsa.SimpleExpSmoothing(y).fit(smoothing_level=0.5)
fitted_ses = model_ses.fittedvalues
predic_ses = model_ses.predict(test[0],test[3])

model_ses = pd.concat([fitted_ses,predic_ses])

fig, ax= plt.subplots(figsize=(15,10))
y.plot(color='red', label='Total de minutos salientes (miles)')
model_ses.plot(color='black', label='Simple exponential smoothing', alpha=0.8)
model_RW.plot(color='green', label='Random Walk', alpha=0.8)
ax.set_xticks([], minor=True) 
plt.legend(loc='best')
plt.show()
```

Comparando ambos modelos se observa que para valores pequeños de h las predicciones resultan mejores usando random walk ya que en puntos cercanos al último punto conocido los valores predecidos son cercanos a éste último.
Para valores de h mayores la predicción usando exponencial smoothing debería ser mejor ya que a diferencia del primero, captura información de periodicidad y estacionalidad, pero en este caso se estabiliza demaciado rápido haciendose constante por los que tampoco resulta una buena predicción.
